- Write a program to display IMDB's Top rated 100 movies data (i.e. Name, IMDB ratings, year release and make a dataframe)

```python
url_ = "https://www.imdb.com/chart/top/"
imdb = requests.get(url_)
imdb
```

```python
s = BeautifulSoup(imdb.content)
s
```

```python
# extracting movie names
names_=[]
movies = s.find_all('td',class_='titleColumn')
for i in movies:
    for j in i.find_all("a"):
        names_.append(j.text.strip())
movie_names=names_[0:100]
```

```python
# extracting year release 
year_=[]
years = s.find_all("span",class_="secondaryInfo")
for i in years:
    j = i.text.replace('(',"")
    year_.append(j.replace(')',""))
year_release=year_[0:100]
```

```python
# extracting ratings 
ratings_ = []
ratings = s.find_all("td",class_="ratingColumn imdbRating")
for i in ratings:
    ratings_.append(i.text.strip())
imdb_ratings=ratings_[0:100]
```

```python
#creating dataframe for top 100 movies

Imdb = pd.DataFrame({})
Imdb["Movie Name"] = movie_names
Imdb["Year of Release"] = year_release
Imdb["Imdb Rating"] = imdb_ratings
Imdb
```

- Write a python program to extract information about the local weather from the National Weather Service website of USA, https://www.weather.gov/ for the city, San Francisco. You need to extract data about 7 day extended forecast display for the city. The data should include period, short description, temperature and description.

```python
weather = "https://forecast.weather.gov/MapClick.php?textField1=37.78&textField2=-122.42#.YLYJoJMzZQI"
response = requests.get(weather)
response
```

```python
soup = BeautifulSoup(response.content)
```

```python
# extracting period name
period =[]
for i in soup.find_all("p",class_="period-name"):
    period.append(i.text)
```

```python
# extracting short description
short_desc = []
for i in soup.find_all("p",class_="short-desc"):
    short_desc.append(i.text)
```

```python
# extracting temprature
low=[]
high=[]
for i in soup.find_all("p",class_="temp temp-low"):
    low.append(i.text)
for j in soup.find_all("p",class_="temp temp-high"):
    high.append(j.text)
```

```python
temperature=[]
from itertools import chain
zipped = list(chain.from_iterable(zip(low, high)))
od = len(low)
ev = len(high)
if od == ev:
    temperature.extend(zipped)
elif od > ev:
    temperature.extend(zipped + low[ev-od:])
else:
    temperature.extend(zipped + high[od-ev:])
```

```python
# extracting detailed description
det_desc =[]
for i in soup.find_all("div",class_="col-sm-10 forecast-text"):
    det_desc.append(i.text)
```

```python
# creating dataframe

Forecast = pd.DataFrame({})
Forecast['Period'] = period
Forecast['Short Description'] = short_desc
Forecast['Temprature'] = temperature
Forecast['Detailed Description'] = det_desc[0:9]
Forecast.index = np.arange(1,len(Forecast)+1)

Forecast
```

- Write a python program to scrape fresher job listings from ‘https://internshala.com/’. It should include job title, company name, CTC, and apply date.

```python
jobs_url = "https://internshala.com/fresher-jobs"
res = requests.get(jobs_url)
res
```

```python
soup=BeautifulSoup(res.content)
```

```python
#extracting jobs
jobs = []
for i in soup.find_all("div",class_="heading_4_5 profile"):
    jobs.append(i.text.strip())
```

```python
company = []
for i in soup.find_all("div",class_="heading_6 company_name"):
    company.append(i.text.strip())
```

```python
# extracting job location
location =[]
for i in soup.find_all("p",id="location_names"):
    location.append(i.text.strip())
```

```python
ctc = []
lst=[]
for i in soup.find_all("div",class_="item_body"):
    lst.append(i.text.strip())
    
ctc = [x for x in lst if "LPA" in x]
```

```python
#Extracting apply date from prevoius list
apply_date = [x for x in lst if "' 21" in x]
```

```python
#Creating a dataframe
import pandas as pd
import numpy as np

internshala_jobs = pd.DataFrame({})
internshala_jobs['Job Title'] = jobs
internshala_jobs['Company Name'] = company
internshala_jobs['Job Location'] = location
internshala_jobs['Salary Offered'] = ctc
internshala_jobs['Apply Date'] = apply_date
internshala_jobs.index = np.arange(1,len(internshala_jobs)+1)

internshala_jobs
```
