In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a dataframe and csv. In case if any product has less than 3 pages in search results then scrape all the products available under that product vertical. Details to be scraped are: "Brand Name", "Name of the Product", "Rating", "No. of Ratings", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Other Details" and “Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

```python
driver = webdriver.Chrome('/Users/abhishekmishra/chromedriver')
driver.maximize_window()
driver.get("https://www.amazon.in/")
driver.implicitly_wait(10)

inp = input("Enter an item : ")
search = driver.find_element_by_xpath("//input[@id='twotabsearchtextbox']")
search.clear()
search.send_keys(inp)
button = driver.find_element_by_xpath("//input[@id='nav-search-submit-button']")
button.click()
```

```python
Brand_Name = []
Product_Name = []
Rating = []
No_of_Ratings = []
Price = []
Return_Exchange = []
Expected_Delivery= []
Availability = []
Product_URL = []
Other_Details = []
```

```python
for i in range(0,3):
    for j in driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']/span"):
        if j.text == None:
            Product_Name.append("--")
        else:
            Product_Name.append(j.text)
    for k in driver.find_elements_by_xpath("//h2[@class='a-size-mini a-spacing-none a-color-base s-line-clamp-2']/a"):
        Product_URL.append(k.get_attribute("href"))
    driver.find_element_by_xpath("//li[@class='a-last']").click()
    driver.implicitly_wait(5)
```

```python
for i in Product_URL:
    driver.get(i)
    try:
        brand = driver.find_element_by_xpath("//td[@class='a-span9']//span")
        Brand_Name.append(brand.text)
    except NoSuchElementException:
        Brand_Name.append("--")
        
    try:
        rating = driver.find_element_by_xpath("//span[@class='a-size-medium a-color-base']")
        Rating.append(rating.text)
    except NoSuchElementException:
        Rating.append("-")

  
    try:
        num_rating = driver.find_element_by_xpath("//span[@id='acrCustomerReviewText']")
        No_of_Ratings.append(num_rating.text)
    except NoSuchElementException:
        No_of_Ratings.append("-")
        
        
    try:
        price = driver.find_element_by_xpath("//span[@class='a-size-medium a-color-price priceBlockBuyingPriceString']") 
        Price.append(price.text)
    except NoSuchElementException:
        Price.append("-")
        
     
    try:
        ret_exch = driver.find_element_by_xpath("//span[@data-action='a-popover']/div/a") 
        Return_Exchange.append(ret_exch.text)
    except NoSuchElementException:
        Return_Exchange.append("-")
        
    
    try:
        delivery = driver.find_element_by_xpath("//div[@class='a-section a-spacing-mini']/b") 
        Expected_Delivery.append(delivery.text)
    except NoSuchElementException:
        Expected_Delivery.append("-")
    
    
    try:
        avail=driver.find_element_by_id("availability")
        Availability.append(avail.text)
    except NoSuchElementException:
        Availability.append("-")
        
    try:
        details = driver.find_element_by_id('feature-bullets')
        Other_Details.append(details.text)
    except NoSuchElementException:
        Other_Details.append("-")

  
driver.close()
```

```python
amazon=pd.DataFrame({})
amazon['Brand']=Brand_Name 
amazon['Product']=Product_Name
amazon['Ratings']=Rating
amazon['No of Ratings']=No_of_Ratings
amazon['Price']=Price
amazon['Return Exchange']=Return_Exchange
amazon['Expected Delivery']=Expected_Delivery
amazon['Availability'] = Availability
amazon['Product url'] = Product_URL
amazon['Other Details'] = Other_Details
amazon.index=np.arange(1,len(amazon)+1)

amazon
```

```python
amazon.to_csv("amazon.csv")
```

Write a program to scrap all the available details of best gaming laptops from digit.in.

```python
driver = webdriver.Chrome('/Users/abhishekmishra/chromedriver')
driver.maximize_window()
driver.get("https://www.digit.in/top-products/best-budget-gaming-laptop-3591.html")
driver.implicitly_wait(10)

Model_Name=[]
OS=[]
processor=[]
memory=[]
dimensions=[]
gpu =[]
weight=[]
price=[]
display=[]

for i in driver.find_elements_by_xpath('//div[@class="TopNumbeHeading active sticky-footer"]/a'):
    Model_Name.append(i.text)
    
for j in driver.find_elements_by_xpath("//div[@class='Spcs-details']/table/tbody/tr[3]/td[3]"):
    OS.append(j.text)

for k in driver.find_elements_by_xpath("//div[@class='Spcs-details']/table/tbody/tr[5]/td[3]"):
    processor.append(k.text)
    
for l in driver.find_elements_by_xpath("//div[@class='Spcs-details']/table/tbody/tr[6]/td[3]"):
    memory.append(l.text)
    
for m in driver.find_elements_by_xpath("//div[@class='Spcs-details']/table/tbody/tr[7]/td[3]"):
    dimensions.append(m.text)
    
for n in driver.find_elements_by_xpath("//div[@class='Spcs-details']/table/tbody/tr[8]/td[3]"):
    gpu.append(n.text)
    
for o in driver.find_elements_by_xpath("//div[@class='Spcs-details']/table/tbody/tr[9]/td[3]"):
    weight.append(o.text)
    
for p in driver.find_elements_by_xpath("//td[@class='smprice']"):
    price.append(p.text)
    
for q in driver.find_elements_by_xpath("//div[@class='Spcs-details']/table/tbody/tr[4]/td[3]"):
    display.append(q.text)
    
    
driver.close()
```

```python
laptops = pd.DataFrame()
laptops['Model name'] = Model_Name
laptops['Operating System'] = OS
laptops['Display'] = display
laptops['Processor'] = processor
laptops['Memory'] = memory
laptops['Dimensions'] = dimensions
laptops['Graphics Processor'] = gpu
laptops['Price'] = price

laptops.index=np.arange(1,len(laptops)+1)

laptops
```

Write a python program to scrape the details for all billionaires from www.forbes.com. Details to be scrapped: “Rank”, “Name”, “Net worth”, “Age”, “Citizenship”, “Source”,
“Industry”.

```python
driver = webdriver.Chrome('/Users/abhishekmishra/chromedriver')
driver.maximize_window()
driver.get("https://www.forbes.com/billionaires/")
driver.implicitly_wait(10)

rank = []
name = []
net_worth = []
age = []
citizenship = []
source = []
industry = []


for i in driver.find_elements_by_xpath("//div[@class='rank']"):
    rank.append(i.text)
for j in driver.find_elements_by_xpath("//div[@class='personName']"):
    name.append(j.text)
for k in driver.find_elements_by_xpath("//div[@class='netWorth']/div"):
    net_worth.append(k.text)
for l in driver.find_elements_by_xpath("//div[@class='age']"):
    age.append(l.text)
for m in driver.find_elements_by_xpath("//div[@class='countryOfCitizenship']"):
    citizenship.append(m.text)
for n in driver.find_elements_by_xpath("//div[@class='source-column']"):
    source.append(n.text)
for o in driver.find_elements_by_xpath("//div[@class='category']/div"):
    industry.append(o.text)
    
driver.close()
```

```python
forbes = pd.DataFrame()
forbes['Rank'] = rank
forbes['Name']=name
forbes['Net Worth']=net_worth
forbes['Age']=age
forbes['Citizenship']=citizenship
forbes['Source'] = source
forbes['Industry'] = industry
forbes.index=np.arange(1,len(forbes)+1)

forbes
```
