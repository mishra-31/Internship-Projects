- Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10 jobs data.


```python
opt = Options()
opt.add_argument("--headless")
driver = webdriver.Chrome('/Users/abhishekmishra/chromedriver',options=opt)
driver.maximize_window()
driver.get("https://www.naukri.com/")
driver.implicitly_wait(10)

driver.find_element_by_xpath("//input[contains(@id,'qsb-keyword-sugg')]").send_keys("Data Analyst") 
driver.find_element_by_xpath("//input[contains(@id,'qsb-location-sugg')]").send_keys("Bangalore")
driver.find_element_by_xpath("//button[text()='Search']").click()

job_title = []
for i in driver.find_elements_by_xpath("//a[contains(@class,'title fw500 ellipsis')]"):
    job_title.append(i.text)


company_name = []
for i in driver.find_elements_by_xpath("//a[contains(@class,'subTitle ellipsis fleft')]"):
    company_name.append(i.text)


job_location = []
for i in driver.find_elements_by_xpath("//li[contains(@class,'fleft grey-text br2 placeHolderLi location')]"):
    job_location.append(i.text)


experience = []
for i in driver.find_elements_by_xpath("//li[contains(@class,'fleft grey-text br2 placeHolderLi experience')]"):
    experience.append(i.text)
driver.close()

################################################################
naukri_com = pd.DataFrame({})
naukri_com['Job Title']=job_title
naukri_com['Company']=company_name
naukri_com['Location']=job_location
naukri_com['Experience Required']=experience

naukri_com.index = np.arange(1,len(naukri_com)+1)
naukri_com
```

-  Q2: Write a python program to scrape data for “Data Scientist” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name, full job-description. You have to scrape first 10 jobs data.

```python

driver = webdriver.Chrome('/Users/abhishekmishra/chromedriver')

driver.maximize_window()
driver.get("https://www.naukri.com/")
driver.implicitly_wait(10)

driver.find_element_by_xpath("//input[contains(@id,'qsb-keyword-sugg')]").send_keys("Data Scientist") 
driver.find_element_by_xpath("//input[contains(@id,'qsb-location-sugg')]").send_keys("Bangalore")
driver.find_element_by_xpath("//button[text()='Search']").click()

title=[]
location=[]
company=[]
jd=[]
url = [] 
for i in driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']"):
    url.append(i.get_attribute("href"))


for i in url[:10]:
    driver.get(i)
    driver.implicitly_wait(3)
    
    try:
        job=driver.find_element_by_xpath("//h1[@class='jd-header-title']")
        title.append(job.text)
    except NoSuchElementException:
        pass
    
    try:
        job1=driver.find_element_by_xpath("//h1[@class='av-special-heading-tag ']")
        title.append(job1.text)
    except NoSuchElementException:
        pass
    
    try:
        loc=driver.find_element_by_xpath("//span[contains(@class,'loc')]")
        location.append(loc.text)
    except NoSuchElementException:
        location.append("NA")
    
    try:
        com=driver.find_element_by_xpath("//p[@class='cpName f14']")
        company.append(com.text)
    except NoSuchElementException:
        pass
    
    
    try:
        com1=driver.find_element_by_xpath("//a[@class='pad-rt-8']")
        company.append(com1.text)
    except NoSuchElementException:
        pass
    
    try:
        desc=driver.find_element_by_xpath("//section[contains(@class,'getJobDescription JD')]")
        jd.append(desc.text)
    except NoSuchElementException:
        pass
    
    
    try:
        desc1=driver.find_element_by_xpath("//div[@class='dang-inner-html']")
        jd.append(desc1.text)
    except NoSuchElementException:
        pass
    
driver.close() 
```

```python
jobs=pd.DataFrame({})
jobs['Title']=title
jobs['Location']=location
jobs['Company']=company
jobs['Job Description']=jd
jobs.index=np.arange(1,len(jobs)+1)
jobs
```

- Q6 : Scrape data of first 100 sunglasses listings on flipkart.com. You have to scrape four attributes:
1. Brand
2. Product Description
3. Price
4. Discount %

```python
driver = webdriver.Chrome('/Users/abhishekmishra/chromedriver')
driver.maximize_window()
driver.get("https://www.flipkart.com/")
driver.implicitly_wait(10)
driver.find_element_by_xpath("//button[@class='_2KpZ6l _2doB4z']").click()

driver.find_element_by_xpath("//input[contains(@title,'Search for products, brands and more')]").send_keys("sunglasses")
driver.find_element_by_xpath("//button[contains(@type,'submit')]").click()

url=[]
for i in driver.find_elements_by_xpath("//div[@class='_2MImiq']//a"):
    url.append(i.get_attribute("href"))
    
brand=[]
for i in url:
    driver.get(i)
    driver.implicitly_wait(3)
    for j in driver.find_elements_by_xpath("//div[@class='_2WkVRV']"):
        brand.append(j.text)
        
product=[]
for i in url:
    driver.get(i)
    driver.implicitly_wait(3)
    for j in driver.find_elements_by_xpath("//a[@class='IRpwTa']"):
        product.append(j.text)
        
price=[]
for i in url:
    driver.get(i)
    driver.implicitly_wait(3)
    for j in driver.find_elements_by_xpath("//div[@class='_30jeq3']"):
        price.append(j.text)
        
        
discount=[]
for i in url:
    driver.get(i)
    driver.implicitly_wait(3)
    for j in driver.find_elements_by_xpath("//div[@class='_3Ay6Sb']//span"):
        discount.append(j.text)
        
driver.close()
```

```python
sunglasses=pd.DataFrame({})
sunglasses['Brand']=brand[:100]
sunglasses['Product']=product[:100]
sunglasses['Price']=price[:100]
sunglasses['Discount']=discount[:100]
sunglasses.index=np.arange(1,len(sunglasses)+1)

sunglasses
```

As shown in the above page you have to scrape the tick marked attributes. These are
1. Rating
2. Review_summary
3. Full review

You have to scrape this data for first 100 reviews.

```python
driver = webdriver.Chrome('/Users/abhishekmishra/chromedriver')
driver.maximize_window()
url="https://www.flipkart.com/apple-iphone-11-black-64-gb-includes-earpods-power-adapter/p/itm0f37c2240b217?pid=MOBFKCTSVZAXUHGR&lid=LSTMOBFKCTSVZAXUHGREPBFGI&marketplace"
driver.get(url)
driver.implicitly_wait(10)

driver.find_element_by_xpath("//div[@class='_3UAT2v _16PBlm']").click()

rating=[]
review_summ=[]
review_full=[]


urls=[]
for i in driver.find_elements_by_xpath("//nav[@class='yFHi8N']//a"):
    urls.append(i.get_attribute("href"))

for i in urls:
    driver.get(i)
    driver.implicitly_wait(4)
    
    for j in driver.find_elements_by_xpath("//div[@class='_3LWZlK _1BLPMq']"):
        rating.append(j.text)
    
    for k in driver.find_elements_by_xpath("//p[@class='_2-N8zT']"):
        review_summ.append(k.text)
        
    for l in driver.find_elements_by_xpath("//div[@class='t-ZTKy']"):
        review_full.append(l.text)

driver.close()

################################################################################# 

iphone = pd.DataFrame({})
iphone['Rating']=rating[:100]
iphone['Review_summary']=review_summ[:100]
iphone['Full review']=review_full[:100]

iphone.index=np.arange(1,len(iphone)+1)
iphone
```

- Q10: Go to webpage https://www.amazon.in/
  Enter “Laptop” in the search field and then click the search icon.
  Then set CPU Type filter to “Intel Core i7” and “Intel Core i9”
  After setting the filters scrape first 10 laptops data. You have to scrape 3 attributes for     each laptop:
1. title
2. Ratings
3. Price

```python
driver = webdriver.Chrome('/Users/abhishekmishra/chromedriver')

driver.maximize_window()
driver.get("https://www.amazon.in/")
driver.implicitly_wait(10)
driver.find_element_by_xpath("//input[contains(@id,'twotabsearchtextbox')]").send_keys("Laptop")
driver.find_element_by_xpath("//div[contains(@class,'nav-search-submit nav-sprite')]").click()
driver.find_element_by_xpath("//span[text()='Intel Core i7']").click()
driver.find_element_by_xpath("//span[text()='Intel Core i9']").click()

title=[]
rating=[]
price=[]

for t in driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']"):
    if t.text == None:
        title.append("NA")
    else:
        title.append(t.text)
    
    
for p in driver.find_elements_by_xpath("//span[@class='a-price']"):
    if p.text == None:
        price.append("NA")
    else:
        price.append(p.text)

url=[]    
for i in driver.find_elements_by_xpath("//a[@class='a-size-base a-link-normal a-text-normal']"):
    url.append(i.get_attribute("href"))
rating=[]
for i in url:
    driver.get(i)
    driver.implicitly_wait(4)
    
    try:
        r = driver.find_element_by_xpath("//span[@class='a-size-medium a-color-base']")
        rating.append(r.text)
    except NoSuchElementException:
        rating.append("NA")
        
driver.close()      
   ################################################################################# 

laptops = pd.DataFrame({})
laptops['Name']=title[:10]
laptops['Rating']=rating[:10]
laptops['Price']=price[:10]

laptops.index=np.arange(1,len(laptops)+1)
laptops
```
