# Deccan Herald Scrapper

```python
import numpy as np
import pandas as pd 
from selenium import webdriver
from bs4 import BeautifulSoup
import requests
```

```python
driver = webdriver.Chrome('/Users/abhishekmishra/chromedriver')
driver.maximize_window()
news_url = []
for i in range(1,32):
    driver.get("https://www.deccanherald.com/archives")
    driver.implicitly_wait(10)
    date = driver.find_element_by_xpath('//*[@id="arc_datepicker"]')  
    date.click()
    driver.find_element_by_xpath('/html/body/div[2]/div[1]/div/div/span[2]').click()
    driver.find_element_by_xpath(f'//span[contains(@aria-label,"August {i}, 2020")]').click()
    driver.find_element_by_xpath('//a[text()="Submit"]').click()
    for i in driver.find_elements_by_xpath('//li[@class="group sanspro-reg archives-note"]/a'):
        news_url.append(i.get_attribute("href"))
```

```python
Date = []
Vertical = []
Author = []
Headline = []
Description = []
```

```python
for i in news_url:
    driver.get(i)
    
    driver.implicitly_wait(5)
    
    try:
        date = driver.find_element_by_xpath('//ul[@class="crud-items"]/li')
        Date.append(date.text)
    except NoSuchElementException:
        Date.append("-")
        
    try:
        vertical = driver.find_element_by_xpath('//div[@class="item-list"]/ul/li[2]/a')
        Vertical.append(vertical.text)
    except NoSuchElementException:
        Vertical.append("-")
        
        
    try:
        headline = driver.find_element_by_xpath('//h1[@id="page-title"]')
        Headline.append(headline.text)
    except NoSuchElementException:
        Headline.append("-")
        
    try:
        description = driver.find_element_by_xpath('//div[@class="field field-name-body field-type-text-with-summary field-label-hidden"]')
        Description.append(description.text)
    except NoSuchElementException:
        Description.append("-")
        
        
    try:
        author = driver.find_element_by_xpath('//a[@class="sanspro-reg article-author__name"] | //span[@class="sanspro-reg article-author__name"]')        
        Author.append(author.text)
    except NoSuchElementException:
        Author.append("-")
```

```python
news = pd.DataFrame()
news['Date'] = Date
news['Vertical'] = Vertical
news['Author'] = Author
news['Headline'] = Headline 
news['Description'] = Description

news.index=np.arange(1,len(news)+1)

news
```

```python
driver.close()
```

```python
news.to_csv("News.csv")
```
