2. Scrape the details team India’s international fixtures from bcci.tv. Url = https://www.bcci.tv/.
You need to find following details:

A) Match title (I.e. 1st ODI)

B) Series

C) Place

D) Date

E) Time

Note: - From bcci.tv home page you have reach to the international fixture page through code.

```python
driver = webdriver.Chrome('/Users/abhishekmishra/chromedriver')
driver.maximize_window()
driver.get("https://www.bcci.tv/")
driver.implicitly_wait(10)

page = driver.find_element_by_xpath("//a[@class='navigation__link navigation__link--in-drop-down']")  
search = page.get_attribute('href')

driver.get(search)
```

```python
match_title=[]
series = []
place = []
date = []
time = []

for m in driver.find_elements_by_xpath('//strong[@class="fixture__name fixture__name--with-margin"]'):
    match_title.append(m.text)

for s in driver.find_elements_by_xpath('//span[@class="u-unskewed-text fixture__tournament-label u-truncated"]'):
    series.append(s.text)
    
for p in driver.find_elements_by_xpath('//p[@class="fixture__additional-info"]/span'):
    place.append(p.text)
    
```

```python
url=[]
for i in driver.find_elements_by_xpath('//div[@class="js-list"]/a'):
    url.append(i.get_attribute('href'))
```

```python
for i in url:
    driver.get(i)
    d=driver.find_element_by_xpath('//div[@class="mc-header-scorebox__datetime"]/strong')
    date.append(d.text)
    
    t=driver.find_element_by_xpath('//span[@class="mc-header-scorebox__ist-time"]')
    time.append(t.text)
```

```python
driver.close()
```

```python
matches = pd.DataFrame()
matches['Match title']=match_title
matches['Series']=series
matches['Place']=place
matches['Date']=date
matches['Time']=time

matches.index = np.arange(1,len(matches)+1)
matches
```

 5. Scrape the details of trending repositories on Github.com. Url = https://github.com/
You have to find the following details:
    
A) Repository title

B) Repository description 

C) Contributors count

D) Language used

Note: - From the home page you have to click on the trending option from Explore menu through code.

```python
driver = webdriver.Chrome('/Users/abhishekmishra/chromedriver')
driver.maximize_window()
driver.get(" https://github.com/")
driver.implicitly_wait(5)
```

```python
driver.find_element_by_xpath('//div[@class="position-relative mr-3"]').click()
driver.find_element_by_xpath("//input[@name='login']").send_keys("mishra-31")
driver.find_element_by_xpath("//input[@name='password']").send_keys("xxxxxxxxxxx")
driver.find_element_by_xpath("//input[@name='commit']").click()
driver.find_element_by_xpath("/html/body/div[1]/header/div[3]/nav/a[4]").click()
driver.find_element_by_xpath('//*[@id="js-pjax-container"]/div[1]/nav/div/a[3]').click()
```

```python
url = []
for i in driver.find_elements_by_xpath('//h1[@class="h3 lh-condensed"]/a'):
    url.append(i.get_attribute('href'))
```

```python
repository_title = []
repository_description = []
contributors_count = []
language_used = []
for i in url:
    driver.get(i)
    driver.implicitly_wait(3)
    
    try:
        title = driver.find_element_by_xpath('//strong[@itemprop="name"]/a')
        repository_title.append(title.text)
    except NoSuchElementException:
        repository_title.append('-')
        
        
    try:
        description = driver.find_element_by_xpath('//p[@class="f4 mt-3"]')
        repository_description.append(description.text)
    except NoSuchElementException:
        repository_description.append('-')
        
    
    try:
        contributors = driver.find_element_by_xpath('//h2[@class="h4 mb-3"]/a[contains(text(),"Contributors")]/span')    
        contributors_count.append(contributors.text)
    except:
        contributors_count.append('-')
        
    l=[]
    languages=driver.find_elements_by_xpath("//li[@class='d-inline']//a//span[1]")
    if languages:
        for i in languages:
            l.append(i.text)
    else:
        l.append("No languages used")
    language_used.append(l)
```

```python
driver.close()

github = pd.DataFrame()
github['Repository']=repository_title
github['Description']=repository_description
github['No of contributors']=contributors_count
github['Languages used']=language_used

github.index=np.arange(1,len(github)+1)
github
```
